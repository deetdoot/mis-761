### Methods

<div style="font-size: 12px;">
&nbsp; &nbsp; Our predictive methods include Classification Techniques of Logistic Regression, K-Nearest Neighbors, and Random Forest, using all predictors except stroke.<br>
&nbsp; &nbsp; The histogram of average glucose level is bimodal, with peaks occurring at about 80 and 220. Normal blood sugar would be 99 mg/dL or lower after a blood test in which the patient fasted overnight (Northeast). A blood sugar of 100 mg/dL or higher is considered abnormal. A blood sugar of 126 mg/dL or higher is considered type 2 diabetes. We create an indicator variable for diabetes. There are 18.6% of observations with diabetes. <br>
&nbsp; &nbsp; Weighted Logistic Regression improves classification performance. The data is imbalanced with 4.26% of observations label having stroke. We use logistic regression weights of 1 divided by proportion of observations having stroke, on observations having stroke. Coefficient on diabetes isn't significant in the logistic regression. Significant predictors include age, hypertension, and heart disease.<br>
$O = odds(stroke = 1) = e^{\beta_0+\beta_1x_1+...}$<br>
Then $e^{\beta_0+c\beta_1x_1+...} = Oe^{c}$<br>
Odds equals probability divided by 1 minus probability. Interpreting the coefficients, a person with hypertension has odds $e^{0.52}=1.68$ times higher of having a stroke than without hypertension. A one year increase in age increases odds by 1.1 times. Having heart disease increases odds by 1.4 times.
A one unit increase in avg_glucose level or bmi doesn't affect odds since $e^{\text{coefficient}}\approx 1$.<br>
For weighted logistic regression, the coefficients are age: 0.08, hypertension: 0.69, heart_disease: 0.36, avg_glucose_level: 0.00, bmi: 0.01. The corresponding odds factor increases are: age: 1.1 times, hypertension: 2 times, heart_disease: 1.4 times, and no change for one unit increases in avg_glucose_level or bmi.
<br>
&nbsp; &nbsp; Accuracy equals true positives plus true negatives over all predictions. Precision equals true positives over all predicted positives. Specificity equals true negatives over all predicted negatives. Recall/sensitivity equals true positives over all positives.

&nbsp; &nbsp; The choice of K is very important to the performance of KNN (Bruce). The simplest choice is to set K = 1, based on finding the data record in the training set most similar to the new record to be predicted. Setting K = 1 is rarely the best choice; you'll almost always obtain superior performance by using K > 1. Recall and precision decrease to 0 in our case increasing K. <br>
&nbsp; &nbsp; Random Forest is type of bagged estimate based on decision tree models. The random forest is based on applying bagging to decision trees, with one important extension: in addition to sampling the records, the algorithm also samples the variables. The random forest method is a "black box" method.<br>
 &nbsp; &nbsp; We obtain perfect classification in the confusion matrix using Random Forest; all observations are correctly predicted for stroke and not stroke. We analyze variable importance. Using 20% of data for the validation set, measure Random Forest Classifier accuracy, then for each column shuffle its values, and predict using Random Forest Classifier with the column's random ordering, and calculate the percentage decease in accuracy, to measure the variable's importance. The greater the decrease, the more important the variable is for Random Forest Classifier. The most important variable is work_typeSelfEmployed with 0.05% decrease in accuracy; randomly shuffling predictors including age cause accuracy to increase by over 0.05%. None of the variables are important predictors of stroke. In another model context, Random Forest Classifier example has 125% decrease in accuracy for most important predictor (Bruce). <br>
  &nbsp; &nbsp; Especially for imbalanced data where the event of interest is rare, the accuracy metric is horrendously misleading for classification problems (Nield). If a vendor, consultant, or data scientist ever tries to sell you a classification system on claims of accuracy, ask for a confusion matrix. An alternate measure of variable importance could utilize F1 score decrease, including precision and recall.
 <br>
